```python
import pandas as pd
```

```python
# YOU SHOULD IMPLEMENT A SINGLE FUNCTION
# DO NOT CHANGE ITS NAME
# IT SHOULD READ THE DATASET FROM THE FILE ALONG WITH THE NOTEBOOK
# IT SHOULD PERFORM THE APPROPRIATE TRANSFORMATIONS TO THE DATAFRAME
# IT SHOULD RETURN THE TRANSFORMED DATAFRAME

def exercise_01():

    df = pd.DataFrame()
    df = pd.read_csv('nat2019.csv', names = ['year', 'name', 'gender', 'births'], delimiter = ';', low_memory=False)
    col_list = list(df)
    col_list[0],col_list[2] = col_list[2],col_list[0]
    df.columns = col_list
    df = df.reindex(columns = ['year', 'name', 'gender', 'births'])
    df = df.drop(index=0)
    df = df.dropna()

    # unusable rows out
    df = df[~df.eq('XXXX').any(1)]

    # cast
    df['year'] = df['year'].astype(int)
    df['name'] = df['name'].astype(str)
    df['gender'] = df['gender'].astype(str)
    df['births'] = df['births'].astype(int)

    # change gender
    df['gender'] = df['gender'].replace(['1','2'],['M','F'])

    # single character out and special charachters minus - out
    df = df.drop(df[df['name'].str.len() == 1].index)
    df = df.drop(df[df['name'].str.contains(r'[^0-9a-zA-Z.-´]')].index)

    # capitalize first letter
    df['name'] = df['name'].str.title()

    # Sorted data
    df = df.sort_values(by = ['year', 'gender','births', 'name'], ascending = [True, True, False, True])
    
    # index reseted dataframe
    df = df.reset_index(drop=True)
    
    return df
```

```python
exercise_01()
```
